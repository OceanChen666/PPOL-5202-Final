```python
import pandas as pd
import numpy as np
```

```python
## Data Reading
production = pd.read_csv("../Data/oil-production-by-country.csv")
production.head()
```

## OPEC Group up

```python
opec_list_1960 = ["Iran","Iraq","Kuwait","Saudi Arabia","Venezuela"]
opec_list_1961 = ["Iran","Iraq","Kuwait","Saudi Arabia","Venezuela","Qatar"]
opec_list_1962 = ["Iran","Iraq","Kuwait","Saudi Arabia","Venezuela","Qatar",
                  "Indonesia","Libya"]
opec_list_1967 = ["Iran","Iraq","Kuwait","Saudi Arabia","Venezuela","Qatar",
                  "Indonesia","Libya","United Arab Emirates"]
opec_list_1969 = ["Iran","Iraq","Kuwait","Saudi Arabia","Venezuela","Qatar",
                  "Indonesia","Libya","United Arab Emirates","Algeria"]
opec_list_1971 = ["Iran","Iraq","Kuwait","Saudi Arabia","Venezuela","Qatar",
                  "Indonesia","Libya","United Arab Emirates","Algeria",
                  "Nigeria"]
opec_list_1973 = ["Iran","Iraq","Kuwait","Saudi Arabia","Venezuela","Qatar",
                  "Indonesia","Libya","United Arab Emirates","Algeria",
                  "Nigeria","Ecuador"]
opec_list_1975 = ["Iran","Iraq","Kuwait","Saudi Arabia","Venezuela","Qatar",
                  "Indonesia","Libya","United Arab Emirates","Algeria",
                  "Nigeria","Ecuador","Gabon"]
opec_list_1992 = ["Iran","Iraq","Kuwait","Saudi Arabia","Venezuela","Qatar",
                  "Indonesia","Libya","United Arab Emirates","Algeria",
                  "Nigeria","Gabon"]
opec_list_1995 = ["Iran","Iraq","Kuwait","Saudi Arabia","Venezuela","Qatar",
                  "Indonesia","Libya","United Arab Emirates","Algeria",
                  "Nigeria"]
opec_list_2007 = ["Iran","Iraq","Kuwait","Saudi Arabia","Venezuela","Qatar",
                  "Indonesia","Libya","United Arab Emirates","Algeria",
                  "Nigeria","Ecuador","Angola"]
opec_list_2009 = ["Iran","Iraq","Kuwait","Saudi Arabia","Venezuela","Qatar",
                  "Libya","United Arab Emirates","Algeria",
                  "Nigeria","Ecuador","Angola"]
opec_list_2017 = ["Iran","Iraq","Kuwait","Saudi Arabia","Venezuela","Qatar",
                  "Libya","United Arab Emirates","Algeria",
                  "Nigeria","Ecuador","Gabon","Angola","Equatorial Guinea"]
opec_list_2018 = ["Iran","Iraq","Kuwait","Saudi Arabia","Venezuela","Qatar",
                  "Libya","United Arab Emirates","Algeria",
                  "Nigeria","Ecuador","Gabon","Angola","Equatorial Guinea",
                  "Congo"]
opec_list_2020 = ["Iran","Iraq","Kuwait","Saudi Arabia","Venezuela",
                  "Libya","United Arab Emirates","Algeria",
                  "Nigeria","Gabon","Angola","Equatorial Guinea",
                  "Congo"]
```

```python
def decide_opec_list(year):
    
    if year <= 1960:
        opec_list = opec_list_1960
    elif year <= 1961:
        opec_list = opec_list_1961
    elif year <= 1962:
        opec_list = opec_list_1962
    elif year <= 1967:
        opec_list = opec_list_1967
    elif year <= 1969:
        opec_list = opec_list_1969
    elif year <= 1971:
        opec_list = opec_list_1971
    elif year <= 1973:
        opec_list = opec_list_1973
    elif year <= 1975:
        opec_list = opec_list_1975
    elif year <= 1992:
        opec_list = opec_list_1992
    elif year <= 1995:
        opec_list = opec_list_1995
    elif year <= 2007:
        opec_list = opec_list_2007
    elif year <= 2009:
        opec_list = opec_list_2009
    elif year <= 2017:
        opec_list = opec_list_2017
    elif year <= 2018:
        opec_list = opec_list_2018
    else:
        opec_list = opec_list_2020
    return opec_list
```

```python
year_list = range(1960,2023)

before_1960 = pd.DataFrame()
for year in range(1900,1960):
    unit = production[production["Year"] == year].reset_index(drop=True)
    ## Compute Share
    world = unit[unit["Entity"] == "World"].reset_index(drop=True)
    unit["Share"] = unit["Oil production (TWh)"] / world["Oil production (TWh)"].sum()
    before_1960 = pd.concat([before_1960,unit],axis=0)

after_1960 = pd.DataFrame()
for year in year_list:
    
    unit = production[production["Year"] == year].reset_index(drop=True)
    
    ## Extract Opec
    opec_list = decide_opec_list(year)
    opec = unit[unit["Entity"].isin(opec_list)].reset_index(drop=True)

    opec_row = pd.DataFrame({"Entity":["OPEC"],"Code":["OPEC"],"Year":[year],"Oil production (TWh)":[opec["Oil production (TWh)"].sum()]})
    unit = unit.append(opec_row,ignore_index=True)

    ## Compute Share
    world = unit[unit["Entity"] == "World"].reset_index(drop=True)
    unit["Share"] = unit["Oil production (TWh)"] / world["Oil production (TWh)"].sum()

    after_1960 = pd.concat([after_1960,unit],axis=0)

production_v1 = pd.concat([before_1960,after_1960],axis=0).reset_index(drop=True)
```

```python
print(production_v1.shape)
print(production.shape)

production_v1.to_csv("../Data/oil-production-by-country-v1.csv",index=False)
```
